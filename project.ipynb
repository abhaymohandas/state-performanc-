# **Data Visualization**

```python

plt.figure(figsize=(30, 8))

plt.xlabel('States', fontsize=14)
plt.ylabel('Scores', fontsize=14)
plt.title('StateWise Report of Infrastructure Quality', fontsize=20, fontweight='bold', color='k')

x_axis = np.arange(len(df['State_wise_scores_of_individual_parameters']))
plt.xticks(x_axis + 0.15, df['State_wise_scores_of_individual_parameters'], rotation=45)

plt.bar(x_axis, df['Quality_of_Road_Infrastructure'], color='#FF6347', label='Road Quality', width=0.4)
plt.bar(x_axis + 0.3, df['Quality_of_Rail_Infrastructure'], color='#32CD32', label='Rail Quality', width=0.4)

plt.grid(True, axis='y', linestyle='--')
plt.legend(loc='upper center', fontsize=10, shadow=True, fancybox=True, title='Infrastructure Types')
plt.show()
```

```python
plt.figure(figsize=(30, 8))
plt.xlabel('States', fontsize=14)
plt.ylabel('Scores', fontsize=14)
plt.title('StateWise Report of Warehousing and Logistics Services Quality', fontsize=20, fontweight='bold', color='k')

x_axis = np.arange(len(df['State_wise_scores_of_individual_parameters']))
plt.xticks(x_axis + 0.15, df['State_wise_scores_of_individual_parameters'], rotation=45)

plt.bar(x_axis, df['Quality_of_Warehousing_Infrastructure'], color='#1E90FF', label='Warehousing Quality', width=0.4)
plt.bar(x_axis + 0.3, df['Quality_of_Logistics_Services'], color='#FFA500', label='Quality_of_Logistics_Services', width=0.4)

plt.grid(True, axis='y', linestyle='--')
plt.legend(loc='upper center', fontsize=10, shadow=True, fancybox=True, title='Quality Types')
plt.show()
```

```python
top10_safety_head = df.sort_values(by=['Safety_and_Security_during_Transportation'], ascending=False).head(10)
top10_safety_tail = df.sort_values(by=['Safety_and_Security_during_Transportation'], ascending=False).tail(10)


fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].pie(top10_safety_head['Safety_and_Security_during_Transportation'], labels=top10_safety_head['State_wise_scores_of_individual_parameters'], autopct='%1.1f%%', startangle=90)
axes[0].set_title('Top 10 States by Transportation Safety', fontweight='bold', color='k')
axes[0].axis('equal')

axes[1].pie(top10_safety_tail['Safety_and_Security_during_Transportation'], labels=top10_safety_tail['State_wise_scores_of_individual_parameters'], autopct='%1.1f%%', startangle=90)
axes[1].set_title('Last 10 States by Transportation Safety', fontweight='bold', color='k')
axes[1].axis('equal')

plt.tight_layout()
plt.show()
```

```python
top10_safety_head = df.sort_values(by=['Efficiency_of_Regulatory '], ascending=False).head(10)
top10_safety_tail = df.sort_values(by=['Efficiency_of_Regulatory '], ascending=False).tail(10)


fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].pie(top10_safety_head['Efficiency_of_Regulatory '], labels=top10_safety_head['State_wise_scores_of_individual_parameters'], autopct='%1.1f%%', startangle=90)
axes[0].set_title('Top 10 States by Efficiency', fontweight='bold', color='k')
axes[0].axis('equal')

axes[1].pie(top10_safety_tail['Efficiency_of_Regulatory '], labels=top10_safety_tail['State_wise_scores_of_individual_parameters'], autopct='%1.1f%%', startangle=90)
axes[1].set_title('Last 10 States by Efficiency', fontweight='bold', color='k')
axes[1].axis('equal')

plt.tight_layout()
plt.show()
```

```python

states = df['State_wise_scores_of_individual_parameters']
road_quality = df['Quality_of_Road_Infrastructure']
rail_quality = df['Quality_of_Rail_Infrastructure']
efficiency = df['Efficiency_of_Regulatory ']
safety = df['Safety_and_Security_during_Transportation']
warehouse = df['Quality_of_Warehousing_Infrastructure']
logistics = df['Quality_of_Logistics_Services']

plt.figure(figsize=(12, 6))
plt.plot(states, road_quality, marker='o', label='Road Quality')
plt.plot(states, rail_quality, marker='o', label='Rail Quality')

plt.plot(states, efficiency, marker='o', label='Efficiency')
plt.plot(states, safety, marker='o', label='safety')

plt.plot(states, warehouse, marker='o', label='warehouse')
plt.plot(states, logistics, marker='o', label='logistics')

plt.xlabel('States')
plt.ylabel('Quality Scores')
plt.title("Overview of States")
plt.xticks(rotation=90)
plt.legend()
plt.grid(True, linestyle='--')

plt.tight_layout()
plt.show()
```
